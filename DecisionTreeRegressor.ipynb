```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
p=pd.read_csv('/content/sample_data/crocodile_dataset.csv')
```

```python
p.head()
```

```python
pd.isna(p).sum()
```

```python
p.max()
```

```python
LabelEncoder().fit_transform(p['Scientific Name'])
```

```python
p['Scientific Name']=LabelEncoder().fit_transform(p['Scientific Name'])
p['Common Name']=LabelEncoder().fit_transform(p['Common Name'])
p['Genus']=LabelEncoder().fit_transform(p['Genus'])
p['Sex']=LabelEncoder().fit_transform(p['Sex'])
p['Country/Region']=LabelEncoder().fit_transform(p['Country/Region'])
p['Habitat Type']=LabelEncoder().fit_transform(p['Habitat Type'])
p['Conservation Status']=LabelEncoder().fit_transform(p['Conservation Status'])
p['Observer Name']=LabelEncoder().fit_transform(p['Observer Name'])
p['Notes']=LabelEncoder().fit_transform(p['Notes'])
p['Age Class']=LabelEncoder().fit_transform(p['Age Class'])
```

```python
p
```

```python
columns=['Scientific Name', 'Family','Date of Observation']
for column in columns:
  p[column]=LabelEncoder().fit_transform(p[column])
print(p)
```

```python
y=p['Observed Weight (kg)']
```

```python
x=pd.DataFrame(p, columns=['Observed Length (m)','Country/Region','Habitat Type'])
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
pd.DataFrame(x,columns=['Observed Length (m)','Country/Region','Habitat Type'])
```

```python
import seaborn as sns
cor_matrix= p.corr()
plt.figure(figsize=(10,10))
sns.heatmap(cor_matrix, annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, test_size=0.25, random_state=42)
```

```python
x_test.shape
from sklearn.linear_model import LinearRegression
Model=LinearRegression()
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)
y_pred
d=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
d
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
Model.coef_
```

```python
Model.intercept_
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Weight (kg)")
plt.ylabel("Predicted Weight (kg)")
plt.title("Actual vs Predicted Weight")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2) # Add a diagonal line for reference
plt.show()
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
Model=DecisionTreeRegressor(max_depth=5, criterion='squared_error')
Model.fit(x_train,y_train)
```

```python
y_pred_dt = Model.predict(x_test)

plt.scatter(y_test, y_pred_dt)
plt.xlabel("Actual Weight (kg)")
plt.ylabel("Predicted Weight (kg)")
plt.title("Actual vs Predicted Weight (Decision Tree Regressor)")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.show()
```

```python
d=pd.DataFrame({'Actual':y_test,'Predicted':y_pred_dt})
d
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_dt)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))
plot_tree(Model, feature_names=['Observed Length (m)','Country/Region','Habitat Type'], filled=True, rounded=True, fontsize=10)
plt.show()
```
