```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
p=pd.read_csv('/content/sample_data/crocodile_dataset.csv')
```

```python
LabelEncoder().fit_transform(p['Scientific Name'])
```

```python
p['Scientific Name']=LabelEncoder().fit_transform(p['Scientific Name'])
p['Common Name']=LabelEncoder().fit_transform(p['Common Name'])
p['Genus']=LabelEncoder().fit_transform(p['Genus'])
p['Sex']=LabelEncoder().fit_transform(p['Sex'])
p['Country/Region']=LabelEncoder().fit_transform(p['Country/Region'])
p['Habitat Type']=LabelEncoder().fit_transform(p['Habitat Type'])
p['Conservation Status']=LabelEncoder().fit_transform(p['Conservation Status'])
p['Observer Name']=LabelEncoder().fit_transform(p['Observer Name'])
p['Notes']=LabelEncoder().fit_transform(p['Notes'])
p['Age Class']=LabelEncoder().fit_transform(p['Age Class'])
```

```python
p
```

```python
columns=['Scientific Name', 'Family','Date of Observation']
for column in columns:
  p[column]=LabelEncoder().fit_transform(p[column])
print(p)
```
